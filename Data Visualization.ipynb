Let's create some basic visualizations using these two libraries

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
# Create sample data: x is an array of 100 values between 0 and 10
x = np.linspace(0, 10, 100) # returns an array of 100 evenly spaced values between 0 and 10 (inclusive).
# y is the sine of x
y = np.sin(x) #returns an array where each element is the sine of the corresponding element in x
y2 = np.cos(x)
# Display a grid to help with readability of the plot
plt.grid(True)
# Show the plot
plt.show()
```

```python
# Create the line plot
plt.plot(x, y, label='Sine Wave', color='blue') # This creates the line plot by connecting the points (x, y) with a line.
plt.plot(x, y2, label='coSine Wave', color='green')
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
# Create sample data: x is an array of 100 values between 0 and 10
x = np.linspace(0, 10, 100) # returns an array of 100 evenly spaced values between 0 and 10 (inclusive).
# y is the sine of x
y = np.sin(x) #returns an array where each element is the sine of the corresponding element in x
y2 = np.cos(x)
# Create the line plot
plt.plot(x, y, label='Sine Wave', color='black') # This creates the line plot by connecting the points (x, y) with a line.
plt.plot(x, y2, label='coSine Wave', color='red')
# Add title to the plot
plt.title('Line Plot - Sine Wave')
# Label the x-axis
plt.xlabel('X-axis')
# Label the y-axis
plt.ylabel('Y-axis')
# Add a legend in the upper right corner
plt.legend()
# Display a grid to help with readability of the plot
plt.grid(True)
# Show the plot
plt.show()
```

```python
# Import necessary libraries
import seaborn as sns
import numpy as np
# Generate random data for the histogram
data = np.random.normal(size=500) # 500 data points sampled from a normal distribution
# normal here means gaussian/normal distribution of data
# Create a histogram with KDE (Kernel Density Estimation) using seaborn
sns.histplot(data, bins=50, kde=True, color='yellow')
# Bin size: refers to the width of each bar, which represents a range of value
# Add a title to the histogram
plt.title('Histogram with KDE')
# Show the plot
plt.show()
```

```python
# Generate example data for the box plot
data = np.random.normal(size=500) # returns an array of 500 random numbers
# Box plot to identify outliers in the 'Age' column
plt.boxplot(data)
plt.title('Box Plot of data')
plt.ylabel('data')
plt.show()
```

```python
import pandas as pd
# Create a dataset for comparison
data = pd.DataFrame({
 'Age': [23, 45, 34, 42, 25, 36, 52, 39, 40, 30],
 'Salary': [40000, 80000, 60000, 75000, 42000, 62000, 95000, 71000,
85000, 53000]
})
# Scatter Plot of Age vs Salary
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')
plt.title('Age vs Salary (Scatter Plot)')
plt.show()
# Histogram of Age
sns.histplot(data['Age'], bins=5, kde=True, color='green')
plt.title('Histogram of Age')
plt.show()
# Box Plot of Salary
sns.boxplot(data['Salary'], color='orange')
plt.title('Box Plot of Salary')
plt.show()
```

Creating all 3 in the same window for comparison

```python
# Create a figure with 1 row and 3 columns
plt.figure(figsize=(15, 5)) # This creates a figure with a specific size, here 15 inches wide and 5 inches tall, to accommodate three plots side by side.
# Scatter Plot (1st subplot)
plt.subplot(1, 3, 1) # (rows, columns, position)
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')
plt.title('Age vs Salary (Scatter Plot)')
plt.xlabel('Age')
plt.ylabel('Salary')
# Histogram (2nd subplot)
plt.subplot(1, 3, 2) # (rows, columns, position)
sns.histplot(data['Age'], bins=5, kde=True, color='green')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
# Box Plot (3rd subplot)
plt.subplot(1, 3, 3) # (rows, columns, position)
sns.boxplot(data['Salary'], color='orange')
plt.title('Box Plot of Salary')
plt.xlabel('Salary')
# Adjust layout to avoid overlap
plt.tight_layout()
# Show the plots
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data
data = pd.DataFrame({
    'Age': [22, 25, 30, 35, 40, 45, 50, 55, 60, 65],
    'Salary': [30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000, 70000, 75000]
})

# Create a figure with 2 rows and 2 columns
plt.figure(figsize=(20, 15))

# Scatter Plot (1st subplot)
plt.subplot(2, 2, 1)
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')
plt.title('Age vs Salary (Scatter Plot)')
plt.xlabel('Age')
plt.ylabel('Salary')

# Histogram (2nd subplot)
plt.subplot(2, 2, 2)
sns.histplot(data['Age'], bins=5, kde=True, color='green')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Box Plot (3rd subplot)
plt.subplot(2, 2, 3)
sns.boxplot(x=data['Salary'], color='orange')
plt.title('Box Plot of Salary')
plt.xlabel('Salary')

# Frequency vs Salary (4th subplot)
plt.subplot(2, 2, 4)
sns.histplot(data['Salary'], bins=5, kde=True, color='blue')
plt.title('Frequency vs Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plots
plt.show()
```
